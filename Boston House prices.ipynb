```python
##Regularization Ridge,lasso and elestic net algorithms which helps to reduce the complexity
from sklearn.linear_model import Ridge,Lasso,ElasticNet
```

```python
rr=Ridge(alpha=0.0025,normalize=True)
rr.fit(x,y)
```

```python
print(rr.coef_)
print("\n",rr.intercept_)

print('\nAccuracy of test data',rr.score(x_test,y_test))
print('Accuracy of train data',rr.score(x_train,y_train))
#MSE and RMSE
print("\nTrain Mean sqr Errors")
print("Train Mean Sqauer errors",mean_squared_error(y_train,rr.predict(x_train)))

print("\nTest Mean sqr Errors")
print("Test Mean Sqauer errors",mean_squared_error(y_test,rr.predict(x_test)))

print("\nTrain Root Mean sqr Errors")
print("Train Root Mean Sqauer errors",sqrt(mean_squared_error(y_train,rr.predict(x_train))))

print("\nTest Root Mean sqr Errors")
print("Test Root Mean Sqauer errors",sqrt(mean_squared_error(y_test,rr.predict(x_test))))
```

```python
#Lasso regression

lass=Lasso(alpha=0.000001,normalize=True)
lass.fit(x,y)
```

```python
print(lass.coef_)
print("\n",lass.intercept_)

print('\nAccuracy of test data',lass.score(x_test,y_test))
print('Accuracy of train data',lass.score(x_train,y_train))
#MSE and RMSE
print("\nTrain Mean sqr Errors")
print("Train Mean Sqauer errors",mean_squared_error(y_train,lass.predict(x_train)))

print("\nTest Mean sqr Errors")
print("Test Mean Sqauer errors",mean_squared_error(y_test,lass.predict(x_test)))

print("\nTrain Root Mean sqr Errors")
print("Train Root Mean Sqauer errors",sqrt(mean_squared_error(y_train,lass.predict(x_train))))

print("\nTest Root Mean sqr Errors")
print("Test Root Mean Sqauer errors",sqrt(mean_squared_error(y_test,lass.predict(x_test))))
```

```python
#elasticnet regression
en=ElasticNet(alpha=0.001,normalize=True)
en.fit(x,y)
```

```python
print(en.coef_)
print("\n",en.intercept_)

print('\nAccuracy of test data',en.score(x_test,y_test))
print('Accuracy of train data',en.score(x_train,y_train))
#MSE and RMSE
print("\nTrain Mean sqr Errors")
print("Train Mean Sqauer errors",mean_squared_error(y_train,en.predict(x_train)))

print("\nTest Mean sqr Errors")
print("Test Mean Sqauer errors",mean_squared_error(y_test,en.predict(x_test)))

print("\nTrain Root Mean sqr Errors")
print("Train Root Mean Sqauer errors",sqrt(mean_squared_error(y_train,en.predict(x_train))))

print("\nTest Root Mean sqr Errors")
print("Test Root Mean Sqauer errors",sqrt(mean_squared_error(y_test,en.predict(x_test))))
```
